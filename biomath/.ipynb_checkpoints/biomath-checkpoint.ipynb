```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
from scipy.optimize import fsolve
```

```python
def f(u, k=2.):
    return u ** 3. * np.exp(k * (1 - u))

def f_diff(u, z=2.):
    return f(u, z) - u

def fu(u, k=2.):
    return u ** 2. * np.exp(k * (1 - u)) * (3 - k * u)

def g(u, z = 2.):
    return u ** 2. * np.exp(z * (1 - u)) - 1
```

```python
u = np.linspace(0, 25, 1000)
z_list = np.linspace(0.4, 5, 16)
# z_list = [0.4 * x for x in range(1, 17)]
# print(z_list)

fig = plt.figure(figsize=(15, 15))

n = len(z_list) // 4
for i, z in enumerate(z_list):
    ax = fig.add_subplot(4, n, i+1)
    ax.set_xlabel('u')
    ax.set_title('z= %.4f' % z)
    ax.plot(u, u, '--', c='blue')

    f1 = f(u, z)
    f2 = f(f1, z)
    f3 = f(f2, z)

    ax.plot(u, f1, c='purple')
    ax.plot(u, f2, c='c')
    ax.plot(u, f3, c='y')
    ax.legend(['u', 'f', '$f^2$', '$f^3$'])

    # u = f(u)
    # при z > 2 корень на отрезке [0, 2/z]
    root1 = 1.
    if z > 2:
        root1 = fsolve(g, 1. / z, args=z)
        ax.set_xlim([-0.1,  2]);
    elif z < 2:
        root1 = fsolve(f_diff, 3./z + 1, args=z)
        ax.set_xlim([-0.1,  5]);   
    if z < 0.8:
        ax.set_xlim([-0.1,  25]);
        
    plt.scatter([0, 1, root1], [0, 1, root1], c='b');

    f1_max = f(3. / z, z); f2_max = f(f1_max, z); f3_max = f(f2_max, z)
    ax.set_ylim([-0.1,  max(f1_max, f2_max, f3_max) + 0.5])
#     ax.set_ylim([-0.1,  2])

#     ax.set_xlim([-0.1,  max(root1, 1) + 1]);
   
    fig.tight_layout()
```

```python
# Особые точки
# u = f(u)

fig = plt.figure(figsize=(8, 5))
ax = fig.gca()
z_list = np.linspace(0.2, 6, 1000)
points = []
for z in z_list:
    root1 = 1.
    if z > 2:
        root1 = fsolve(g, 1. / z, args=z)
    elif z < 2:
        root1 = fsolve(f_diff, 3./z + 1, args=z)
    points.append(root1)
    
ax.plot(z_list, points)
# ax.set_title('Зависимость особой точки от параметра z')
ax.set_xlabel('z')
ax.set_ylabel('$u^*$');

# Устойчивость
fu_point = np.array([np.abs(fu(points[i], z_list[i])) for i in range(len(z_list))])
fig1 = plt.figure(figsize=(8, 5))
ax1 = fig1.gca()

idx_stable = np.where(fu_point < 1)[0]
ax1.plot(points[idx_stable[0]:idx_stable[-1]], fu_point[idx_stable[0]:idx_stable[-1]], c='blue')
ax1.plot(points[:idx_stable[0]], fu_point[:idx_stable[0]], c='red')
ax1.plot(points[idx_stable[-1]:], fu_point[idx_stable[-1]:], c='red')
ax1.legend(['stable points', 'unstable points'])

ax1.set_xlabel('$u^*$')
ax1.set_ylabel('$|f_u(u^*)|$')

ax.plot(z_list[idx_stable[0]:idx_stable[-1]], points[idx_stable[0]:idx_stable[-1]], c='blue')
ax.scatter(z_list[idx_stable[0]], points[idx_stable[0]], c='purple')
ax.scatter(z_list[idx_stable[-1]], points[idx_stable[-1]], c='purple')

ax.legend(['unstable points', 'stable points', 'lost']);
print('Значения z, при которых потеря устойчивости: %.4f, %.4f' % (z_list[idx_stable[0]], z_list[idx_stable[-1]]))
```

```python
# Бифуркационная диаграмма
z_size=1000
z_list = np.linspace(3.1, 5.2, z_size)
n_diag = 150
n_stab = 200
x0 = 2 / z_list
A =  np.zeros((z_size, n_diag))
for i in range(len(z_list)):
    u_cur = x0[i]
    for j in range(n_stab):
        u_cur = f(u_cur, z_list[i])
       
    for j in range(n_diag):
        u_cur =  f(u_cur, z_list[i])
        A[i, j] = u_cur;
        
plt.figure(figsize=(10, 8))
for i, z in enumerate(z_list):
    plt.scatter(z_list[i] * np.ones(n_diag), A[i], s=0.3, c='b')
plt.xlabel('z', fontsize=14, c='black')
plt.ylabel('u', fontsize=14, c='black');
```

```python
def lyap(x0, z):
    n = 1000
    mu = np.zeros(n);
    mu[0] = x0
    for i in range(1, n):
        mu[i] = f(mu[i-1], z)

    mu = sum(np.log(np.abs(fu(mu, z)))) / n
    return mu
```

```python
#  Показатель Ляпунова
z_size=1000
z_list = np.linspace(2.1, 5.2, z_size);
x0 = 2 / z_list
L = np.zeros(z_size)
for i in range(z_size):
    L[i] = lyap(x0[i], z_list[i])

plt.figure(figsize=(10, 8))
plt.scatter(z_list, L, s=1, c='b')
plt.scatter(z_list, np.zeros(z_size), s=1, c='black')
plt.xlabel('z', fontsize=14, c='black')
plt.ylabel('$h(u_1)$', fontsize=14, c='black')
plt.ylim([-7, 1])
```
